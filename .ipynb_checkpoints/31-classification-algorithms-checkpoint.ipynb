```python
#To find the best algorithm for the data
#Data - > Sales data


#import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils.testing import all_estimators
from sklearn import base
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import *
from sklearn.calibration import *
from sklearn.naive_bayes import *
from sklearn.tree.tree import *
from sklearn.gaussian_process.gpc import *
from sklearn.neighbors import *
from sklearn.semi_supervised import * 
from sklearn.discriminant_analysis import *
from sklearn.linear_model import *
from sklearn.neural_network import *
from sklearn.svm import *
from sklearn.feature_selection import RFE
#from sklearn.naive_bayes import *

import importlib


estimators = all_estimators()

class hello:
    
    def __init__(self, X, Y, param):
        X.self = X
        Y.self = Y
        #param.self = param
        
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 42)
        count = 0
        model = []
        arr = []
        for name, class_ in estimators:    
           if issubclass(class_, base.ClassifierMixin): 

            try:
                model = eval(name)()
                model_param = model.get_params()
                
                count = count+1
                #print('\n\n\nAlgorithm No. ', count, name)
                #print(model_param)
               
                #-----------------------Change user defined paramaters-------------------------------------------
                #s={'algorithm': 'SAMME.R', 'base_estimator': None, 'learning_rate': 3, 'n_estimators': 30, 'random_state': None}
                model.set_params(**model_param)
                #print(model.get_params())
                
                #fit model with new changes paramerter
                model_fit = model.fit(X_train, Y_train)
                predicted = model_fit.predict(X_test)
                residual = Y_test - predicted
                #print('predicted',predicted)
                
                #-------------------evauation matrix-----------------------------------
                cm = confusion_matrix(Y_test, predicted)
                #print(name, cm)
                recall = np.diag(cm) / np.sum(cm, axis = 1)
                precision = np.diag(cm) / np.sum(cm, axis = 0)
                np.mean(recall)
                np.mean(precision)
                
                #Accuracy
                diagonal_sum = cm.trace()
                sum_of_all_elements = cm.sum()
                accuracy = diagonal_sum / sum_of_all_elements
                #print('Accuracy :', accuracy)
                #--------------------------------------------------------------------
                
                arr.append([name, accuracy])
                
            except Exception as err:
                print("An exception occurred", err)
                #exit()
        #print(arr)
        df = pd.DataFrame(arr, columns=['Algorithm','Acuuracy'])
        print (df)
```

```python
#Import data using pandas
df = pd.read_csv("data/pima-indians-diabetes.csv")
df.head()


Y = df['class']
X = df.drop(['class'], axis=1)

#s = {'learning_rate': 3.0}
s =''
p1 = hello(X, Y, s)
```
